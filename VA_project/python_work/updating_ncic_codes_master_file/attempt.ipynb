```python
import pandas as pd
import numpy as np

# read in the data
current_dict = pd.read_csv('data/diction_NCICcodes_descriptions_CrimeTiers_CDCTiers_as_csv.csv')
gov_dict = pd.read_csv('data/gov_offical_ncic_list.csv')
```

```python
# in dataframe current_dict, drop the columns that are not needed, 'Notes:', 'Tier', 'CDC Category'
current_dict = current_dict.drop(['Notes:', 'Tier', 'CDC Category'], axis=1)

# in dataframe gov_dict, drop the columns that are not needed, 'Type of Offense Code'
gov_dict = gov_dict.drop(['Type of Offense Code'], axis=1)
```

```python
# convert the column 'NCIC Offense Code' to int 
current_dict['NCIC Offense Code'] = current_dict['NCIC Offense Code'].astype(str)
gov_dict['NCIC Offense Code'] = gov_dict['NCIC Offense Code'].astype(str)
```

```python
# loop through each row in the dataframe 'gov_dict' and check if the value in the column 'NCIC Offense Code' matches any value in the column 'NCIC Offense Code' of dataframe 'current_dict'
# if there is a match, then update new column 'match' to 'yes'
# if there is no match, then update new column 'match' to 'no'

# create new column 'match' in dataframe 'gov_dict'
gov_dict['match'] = np.nan

# loop through each row in dataframe 'gov_dict'
for index, row in gov_dict.iterrows():
    print(row['NCIC Offense Code'])
    print(current_dict['NCIC Offense Code'].values)
    print(row['NCIC Offense Code'] in current_dict['NCIC Offense Code'].values)
    # search the column 'NCIC Offense Code' of dataframe 'current_dict' for the value row['NCIC Offense Code'] and return true if it is found
    if row['NCIC Offense Code'] in current_dict['NCIC Offense Code'].values:
        print('yes')

    # check if the value in the column 'NCIC Offense Code' matches any value in the column 'NCIC Offense Code' of dataframe 'current_dict'
    if row['NCIC Offense Code'] in current_dict['NCIC Offense Code'].values:
        # if there is a match, then update new column 'match' to 'yes'
        gov_dict.loc[index, 'match'] = 'yes'
    else:
        # if there is no match, then update new column 'match' to 'no'
        gov_dict.loc[index, 'match'] = 'no'
```
